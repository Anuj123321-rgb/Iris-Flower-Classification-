```python
def predict_flower(sl, sw, pl, pw):
    # 1. Prepare the data
    data = [[sl, sw, pl, pw]]
    
    # 2. Ask model to predict
    prediction_number = knn.predict(data)[0]
    
    # 3. Translate number to name
    flower_dict = {0: "setosa", 1: "versicolor", 2: "virginica"}
    flower_name = flower_dict[prediction_number]
    
    return flower_name
```

```python
print(predict_flower(5.1, 3.5, 1.4, 0.2))
```

```python
print(predict_flower(6.5, 3.0, 5.2, 2.0))
```

```python
print(predict_flower(9.5, 0.0, 1.2, 1.0))
```
